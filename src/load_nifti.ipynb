## Loading Nifti Format | Niftiファイルの読み込み
nilearnを使ってNiftiファイルを読み込みます．

```python
from nilearn.image import load_img, index_img
from nilearn.plotting import plot_anat
```

### ファイルの読み込み
* nifti_3d : 3次元Nifti画像
* nifti_4d : 4次元Nifti画像

```python
nifti_3d = load_img('data/zstat1.nii.gz')
nifti_4d = load_img('data/filtered_func_data.nii.gz')
```

### 3D画像の出力
[nilearn.plotting.plot_anat](https://nilearn.github.io/modules/generated/nilearn.plotting.plot_anat.html)

```python
plot_anat(nifti_3d)
```

### 4D画像の出力

```python
print(nifti_4d.shape)
```

```python
first_rsn = index_img(nifti_4d, 0)
```

```python
plot_anat(first_rsn)
```
